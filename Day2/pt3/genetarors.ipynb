```python
from typing import Generator, Any
def factorial(n:int) -> Generator[int, Any, str]:
    f=1 
    for i in range(1, n + 1):
        yield f
        f *= i
    return "done"
```

```python
next(factorial(10))
```

```python
from typing import Generator, Any
# also see: the Iterator partern in the design pattern book() (link sourcemaking.com)
def factorial() -> Generator[int, Any, Any]:
    f, i= 1,1
    while True:
        yield f
        f *= i
        i+=1
```

```python
f = factorial()
for _ in range(5):
    print(next(f))
```
