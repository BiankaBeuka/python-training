```python
#LIST: equivalent to the vector in C
#good for adding items and sorting, lenght is not fixed
# disadvantage: slow for calculations
r= list(range(10000000))
r[:10]  # first 10 elements
```

```python
r.pop(0)  # remove first element
%time r.pop(0)  # time to remove first element
```

```python
%time r.pop()  # time to remove last element    
```

```python
# deque (double ended queue) from collections module; mostly like a list, but pop from the beginning is faster
#very fast for adding/removing from both ends
from collections import deque
d= deque(list(range(10000000)))

len(d)
d.append(9999999999999999)
d.popleft()  # remove first element
%time d.popleft()  # time to remove first element
```
