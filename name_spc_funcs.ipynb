```python
def add(a:int, b:int=3, *args, **kwargs) -> int:
    print(f"a={a}, b={b}, args={args}, kwargs={kwargs}")
    return a +b + sum(args) + sum(kwargs.values()) #sum all values from the arguments of the function
add(1, 2)
```
