```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset("penguins")
```

```python
df
```

```python
pd.pivot_table(
   data=df,
   index="island",
   values="bill_length_mm",
   aggfunc="mean"
)
```

```python
pd.pivot_table(
   data=df,
   index="island",
   columns="sex",
   values="bill_length_mm",
   aggfunc="count"
)
```

```python
piv = pd.pivot_table(
   data=df,
   index="island",
   columns="sex",
   values="bill_length_mm",
   aggfunc="count"
)
piv
```

```python
piv.plot.bar()
```
